```python
import pandas as pd
import numpy as np
import requests
import math
```

### People

```python
response = requests.get('https://swapi.dev/api/')
link = response.json()
```

```python
link
```

```python
#people_df.drop(people_df.index, axis=0, inplace=True)
```

```python
response = requests.get(link['people'])
people = response.json()
```

```python
number_of_people = people['count']
next_page = people['next']
previous_page = people['previous']
number_of_results = len(people['results'])
max_page = math.ceil(number_of_people / number_of_results)
```

```python
people_df = pd.DataFrame(people['results'])
```

```python
for i in range(max_page-1):
    response = requests.get(people['next'])
    people = response.json()
    people_df = pd.concat([people_df, pd.DataFrame(people['results'])], ignore_index=True)
```

```python
people_df.tail()
```
